# Pandas Join

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 'B':[10,20,30,]},index=['a','b','c'])

df2=pd.DataFrame({'A':[4,5],
                 'D':[40,50]},index=['a','b'])
display(df1,df2)
```

```python
df1.join(df2,lsuffix='_2')
```

# Append

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                 'B':[10,20,30]})
df2=pd.DataFrame({'c':[1,2,3],
                 'B':[10,20,30]})
display(df1,df2)
```

```python
df1.append(df2)
```

```python
df1.append(df2,ignore_index=True)
```

# Pivot Table

```python
import pandas as pd
```

```python
df=pd.read_csv('Movie_rating.csv')
```

```python
df
```

```python
df.pivot_table(index='Genre')
```

```python
df.pivot_table(index='Genre',columns='Year')
```

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='sum')
```

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='sum',fill_value=0)
```
